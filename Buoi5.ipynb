```python
#Bai 1
def Max_of_two(a, b):
  if a > b:
    return a
  else:
    return b
a, b = map(int, input().split())
print(Max_of_two(a, b))
```

```python
#Bai 2
def swap(a, b):
  return b, a
a, b = map(int, input().split())
print(swap(a, b))
```

```python
#Bai 3
def is_prime(n):
  if n < 2:
    return False
  if n == 2:
    return True
  if n % 2 == 0:
    return False
  for i in range(3, int(n ** 0.5) + 1, 2):
    if n % i == 0:
      return False
  return True
n = int(input())
print(is_prime(n))
```

```python
#Bai 4
def perfect_number(n):
  if n <= 1:
    return False

  tong = 0
  for i in range(1, n // 2 + 1):
    if n % i == 0:
      tong += i
  return tong == n
n = int(input())
print(perfect_number(n))
```

```python
#Bai 5
def find_index(a, k):
  for i in range(len(a)):
    if a[i] == k:
      return i + 1
  return -1

a = list(map(int, input().split()))
k = int(input())
print(find_index(a, k))
```

```python
#Bai 6
def factorial(n):
  if n < 0:
    return None
  if n == 0:
    return 1

  result = 1
  for i in range(1, n + 1):
    result *= i
  return result

n = int(input())
print(factorial(n))
```

```python
#Bai 9
def sum_of_digits(n):
  tong = 0
  while n > 0:
    tong += n % 10
    n //= 10
  return tong

n = int(input())
print(sum_of_digits(n))

```
