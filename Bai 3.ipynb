```python
#Bai 13
def tien_dien(kWh):
  bac1 = 1500
  bac2 = 2000
  bac3 = 3000

  if kWh <= 0:
    return 0

  kWh_1 = min(kWh, 50)
  tien_1 = kWh_1 * bac1
  kWh_2 = max(0, min(kWh,100) - 50)
  tien_2 = kWh_2 * bac2
  kWh_3 = max(0, kWh - 100)
  tien_3 = kWh_3 * bac3

  return int(tien_1 + tien_2 + tien_3)

kWh_input = float(input("Nhap so kWh dien tieu thu: "))
tong_tien = tien_dien(kWh_input)

print(f"Tong tien dien: {tong_tien} vnd.")
```

```python
#Bai 16
def lam_tron_xuong(x):
  if x >= 0:
    return int(x)
  if x == int(x):
    return int(x)
  return int(x) - 1

def lam_tron_len(x):
  if x <= 0:
    return int(x)
  if x == int(x):
    return int(x)
  return int(x) + 1

def lam_tron_gan_nhat(x):
  return int(x + 0.5)

so_thuc = float(input("Nhap vao 1 so thuc: "))
ket_qua_len = lam_tron_len(so_thuc)
ket_qua_xuong = lam_tron_xuong(so_thuc)
ket_qua_gan_nhat = lam_tron_gan_nhat(so_thuc)

print(f"{ket_qua_len} {ket_qua_xuong} {ket_qua_gan_nhat}")




```
