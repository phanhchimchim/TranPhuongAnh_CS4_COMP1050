```python
#Bai 10
def collatz(n):
  count = 1
  while n != 1:
    if n % 2 == 0:
      n //= 2
    else:
      n = 3 * n + 1
    count += 1
  return count
max_length = 0
pos = 0
n = int(input())
for i in range(1, n+1):
  if max_length < collatz(i):
    max_length = collatz(i)
  pos = i
print(pos, max_length)
```

```python
#Bai 11
def uoc_so_chan(n):
  if n % 2 != 0:
    return 0
  count = 0
  for i in range(1, int(n ** 0.5) + 1):
    if n % i == 0:
      if i % 2 == 0:
        count += 1

      j = n // i
      #điều kiện để tránh trường hợp n là scp
      if i != j:
        if j % 2 == 0:
          count += 1
  return count

n = int(input())
if n >= 10 ** 6:
  print("không hợp lệ")

ket_qua = uoc_so_chan(n)
print(ket_qua)


```

```python
#Bai 13
def so_than_thiet(n):
  count = 0
  for i in range(1, int(n ** 0.5) + 1):
    if n % i == 0:
      count += i
      j = n // i
      if i != j:
        count += j

  return count - n

a, b = map(int, input().split())
if a <= 0 or b <= 0:
  print("In ra số nguyên dương.")
else:
  kq_a = so_than_thiet(a)
  kq_b = so_than_thiet(b)
  if a == kq_b and kq_a == b:
    print("true")
  else:
    print("false")

```

```python
#Bai 14
def ucln(m, n):
  while n != 0:
    m, n = n, m % n
  return m

m, n = map(int, input().split())
if m <= 0 or n <= 0:
  print("Nhập số nguyên dương.")
else:
  kq = ucln(m, n)
  print(kq)



```

```python
#Bai 21 cach 2
def tinh_tong(n):
  tong = 0
  so_n = n
  while so_n > 0:
    chu_so_cuoi = so_n % 10
    tong += chu_so_cuoi
    so_n = so_n // 10
  return tong

n = int(input())
print(tinh_tong(n))
```

```python
#Bai 22
def dem(n):
  chu_so = str(n)
  chan = 0
  le = 0
  for i in chu_so:
    if int(i) % 2 == 0:
      chan += 1
    else:
      le += 1
  return chan, le

n = int(input())
print(dem(n))
```
