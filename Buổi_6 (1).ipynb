```python
#W6A4
nums = input().split()
result = {}
for items in nums:
  key, value = items.split(':')
  if key in result:
    result[key].append(value)
  else:
    result[key] = [value]
print(result)
```

```python
#W6A5
nums = list(map(int, input().split()))
result = {'positives': 0, 'negatives': 0, 'zeros': 0}
for n in nums:
  if n > 0:
    result['positives'] += 1
  elif n < 0:
    result['negatives'] += 1
  else:
    result['zeros'] += 1
print(result)
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())
ans = []
for i in range(len(a)):
  for j in range(i + 1, len(a)):
    if a[i] + a[j] == k:
      ans.append((a[i], a[j]))
print(ans)
```

```python
#W6A12
a = list(map(str, input().split()))
a.sort()
d = {}
count = 0
for i in a:
  d[i] = 0
for i in a:
  d[i] += 1
  count = max(count, d[i])
for i in d:
  if d[i] == count:
    print(i)
    break
```

```python
#W6A20

def wev(n: list, k: int) -> int:
    for i, val in enumerate(n):
        if val == k:
            return i + 1
    return -1

a = list(map(int, input().split()))
k = int(input())
print(wev(a, k))
```
